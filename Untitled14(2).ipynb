```python
def DFT(x):
    N = len(x)
    n = np.arange(N, dtype=np.complex64)
    # n = np.arange(N) creates an array of integers from 0 to N-1 for time series index

    k = n.reshape((N, 1))
    # reshapes the array n to  (N, 1) of the form [ [], [], [], [] ],
    # which represents the frequency bins for fourier transform

    e = np.exp(-2j * np.pi * k * n / N)
    # computes the complex exponential term that gets multiplied with the complex signal ,
    #  this is essentially a matrix multiplication

    X = np.dot(e, x)

    # This represents the matrix multiplication of complex matrix e and input signalÂ x
    return X
```

```python
import numpy as np

#modified DFT function for 16000 hz sampling rate which is quite large file

def DFTf(x, blckSize):
    N = len(x)
    X = np.zeros(N, dtype=np.complex64)

    for i in range(0, N, blckSize):
        # converts the signals into smaller blocks
        xBin = x[i:i + blckSize]
        nBin = len(xBin)
        n = np.arange(nBin)

        k = n.reshape((nBin, 1))
        # reshapes the array n to  (N, 1) of the form [ [], [], [], [] ],
        # which represents the frequency bins for fourier transform

        e = np.exp(-2j * np.pi * k * n / nBin)
        # computes the complex exponential term that gets multiplied with the complex signal ,
        #  this is essentially a matrix multiplication

        xBin = np.dot(e, xBin)
        X[i:i + blckSize] = xBin

    return X
```

```python
import librosa
import matplotlib.pyplot as plt

def plotGraph(filePth, samplingRate):
  samplings , samplingRate = librosa.load(filePth , sr = samplingRate , mono = True , offset = 0.0 , duration = None)
  print(samplings)
  X = DFT(samplings)

  # calculate the frequency
  N = len(X)

  n = np.arange(N)
  T = N/samplingRate
  freq = n/T


  plt.figure(figsize = (6, 4))
  plt.stem(freq, abs(X), 'y', \
         markerfmt=" ", basefmt="-b")
  plt.xlabel('Freq (Hz)')
  plt.ylabel('DFT Amplitude |X(freq)|')
  plt.grid()
  plt.show()
  plt.close()

  n_side = N//2
  f_side = freq[:n_side]

  X_side = X[:n_side]/n_side

  plt.subplot(122)
  plt.stem(f_side , abs(X_side) , 'y' , \
           markerfmt = " " , basefmt = "-b")

  plt.xlabel('Freq (Hz)')
  plt.ylabel('Normalised DFT Amplitude |X(freq)|')
  plt.tight_layout()
  plt.grid()
  plt.show()
```

```python
def filteredAudio(file_path, samplingRate):
  import matplotlib.pyplot as plt # Plotting only
  signalSam , sampRate = librosa.load(file_path , sr = samplingRate , mono = True , offset = 0.0 , duration = None)

  def squareWave(x):
    return 0 if x < 0 or x > 1 else 1
  # Generating a sqaure wave

  signAl = [squareWave(x) for x in range(1000)]

  filteredAudio = np.convolve(signalSam, signAl, mode='valid')
  plt.xlabel('Time\nFiltered Signal')
  plt.ylabel('Amplitude')
  plt.grid()
  plt.plot(filteredAudio)
```

```python
def originalAudio(filePath, samplingRate):
  import librosa
  samples , samplingRate = librosa.load(filePath , sr = None , mono = True , offset = 0.0 , duration = None)

  len(samples) , samplingRate

  from librosa import display
  import numpy as np
  import matplotlib.pyplot as plt

  plt.figure()
  librosa.display.waveshow( y = samples, sr = samplingRate)
  plt.show()
```
